```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(1,10); y = np.arange(1,10);
```

```python
[ X, Y ] = np.meshgrid(x,y)
print(X)
```

```python
Z = X*Y
print(Z)
G = np.gradient(Z)
print(G)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, G[0], 1, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
